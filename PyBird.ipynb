```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io.wavfile import read

plt.rcParams['figure.figsize']=(10,6)
```

```python
print("Input path to bird call")
filename = 'C:/Users/liam_/OneDrive/Documents/GitHub/BirdAudio/Data/XC28208 - Eurasian Blue Tit - Cyanistes caeruleus.wav'

fs, data=read(filename)

minval=5000
focussize=int(0.15*fs)
datasize=len(data)
data1=data[150000:190000]
plt.plot(data1)
plt.show()
```

```python
focus=[]
distance=[]

index=0

while index<len(data):
    if data[index]>minval:
        mean_inx=index+focussize//2
        focus.append(float(mean_inx)/datasize)
        if len(focus)>1:
            last_focus = focus[-2]
            actual_focus = focus[-1]
            distance.append(actual_focus-last_focus)
        index+=focussize
    else:
        index+=1
print(focus)
print(distance)


```
